## Section 2

Calculate the historical average returns, standard deviations and correlations.

```python
# Define a function to compute monthly returns.
def compute_monthly_returns(previous: float, current: float)-> float:
    return current / previous - 1
```

### Stock Historical Average Returns and Standard Deviations

```python
# Compute historical average returns, standard deviations and correlations for each stock.
monthly_returns = [] # store the result in a nested list.
for stock in stocks:
    row = []
    stock_arr = stocks_df[stock].values # get the stock price np array
    previous = stock_arr[0]
    for i in range(1, stock_arr.size):
        current = stock_arr[i]
        row.append(compute_monthly_returns(previous, current))
        previous = current
    monthly_returns.append(row)

monthly_returns_df = pd.DataFrame(list(map(list, zip(*monthly_returns))), columns=stocks)
display(monthly_returns_df)
```

```python
print("Historical Average Returns:")
display(monthly_returns_df.mean())
print("Standard Deviation of the Returns:")
display(monthly_returns_df.std())
```

### Monthly Returns of Index Data

```python
sp_monthly_returns = []
previous = sp_df['Adj Close'].values[0]
for i in range(1, len(sp_df['Adj Close'].values)):
    current = sp_df['Adj Close'].values[i]
    sp_monthly_returns.append(compute_monthly_returns(previous, current))
    previous = current
pprint(sp_monthly_returns)
```

### Porfolio Optimization Using Monte Carlo Simulation

```python
num_portfolios = 10000
log_monthly_returns = np.log(stocks_df / stocks_df.shift(1)) # use logarithm for better numerical performance
weights_tangency_porfolio = np.zeros((num_portfolios, len(stocks)))
return_arr = np.zeros(num_portfolios)
std_arr = np.zeros(num_portfolios)
sharpe_arr = np.zeros(num_portfolios)

for portfolio in range(num_portfolios):
    # Get current random weights.
    weights = np.random.randn(len(stocks))
    weights /= np.sum(weights)
    weights_tangency_porfolio[portfolio, :] = weights
    
    # Compute current expected return.
    return_arr[portfolio] = np.sum((np.dot(log_monthly_returns.mean(), weights)) * 12)
    
    # Compute current standard deviation.
    std_arr[portfolio] = np.sqrt(np.dot(weights.T, np.dot(log_monthly_returns.cov() * 12, weights)))
    
    # Sharpe Ratio.
    sharpe_arr[portfolio] = return_arr[portfolio] / std_arr[portfolio]
```

```python
optimal_weights = weights_tangency_porfolio[sharpe_arr.argmax(),:]
expected_return = return_arr[sharpe_arr.argmax()]
standard_deviation = std_arr[sharpe_arr.argmax()]
optimal_sharpe_ratio = sharpe_arr.max()

print('Weights of the tangency portfolio:')
print(optimal_weights)
print(f'Expected return: {expected_return}')
print(f'Standard Deviation: {standard_deviation}')
print(f'Sharpe Ratio: {optimal_sharpe_ratio}')
```

### Generate Visualizations of the Markowitz Bullet.

```python
plt.figure(figsize=(12, 6))
plt.scatter(std_arr, return_arr, c=sharpe_arr, cmap='plasma')
plt.scatter(standard_deviation, expected_return, c='red', s=60, edgecolors='black')
plt.xlim(-5, 60)
plt.ylim(-20, 20)
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Standard Deviation')
plt.ylabel('Return')
```

## Question 6

Find the implied betas, using the tangency portfolio as the market.

```python
# Compute the implied betas.
avg_returns = monthly_returns_df.mean() # compute average monthly return for each stock

betas = dict()
for stock in stocks:
    beta = (avg_returns[stock] - monthly_interest_rate) / (expected_return - monthly_interest_rate)
    betas[stock] = beta

pprint(betas)
```

### Estimate the Historical Beta Values

The historical betas can be estimated as the closed form OLS solution:
$$\mathbf{\beta} = (\mathbf{X'}\mathbf{X})^{-1}\mathbf{X'}\mathbf{Y},$$

where $\mathbf{X}$ is the Index Return and the $\mathbf{Y}$ is the Stock Return.

```python
stock_betas = dict()
Y = np.array(monthly_returns_df.values)
X = np.array(sp_monthly_returns)
XXX = np.dot(1/np.dot(X.T, X), X.T)
beta = np.dot(XXX, Y)
historical_betas = np.column_stack((stocks, beta))
historical_betas_df = pd.DataFrame(historical_betas, columns=['NASDAQ/NYSE CODE', 'Estimated Beta'])
historical_betas_df.set_index('NASDAQ/NYSE CODE', inplace=True)
display(historical_betas_df)
```

```python
# NOTE: the beta can also be computed using OLS regression model provided by statsmodel library, take 3M for example:

X = sp_monthly_returns
Y = monthly_returns_df['MMM']
# X1 = sm.add_constant(X)

# make regression model 
model = sm.OLS(Y, X)

# fit model and print results
results = model.fit()
print(results.summary())
```

### Compute the Expected Returns

```python
# Compute expected returns in percentage form.
expected_return_dict = dict()
risk_premium = 0.06
for stock in stocks:
    beta = float(historical_betas_df['Estimated Beta'][stock])
    expected_return_dict[stock] = (monthly_interest_rate + beta * risk_premium) * 100

expected_return_df = pd.DataFrame(expected_return_dict.items(), columns=['NASDAQ/NYSE CODE', 'Expected Return'])
expected_return_df.set_index('NASDAQ/NYSE CODE', inplace=True)
pd.options.display.float_format = '{:,.2f}%'.format
display(expected_return_df)
```
